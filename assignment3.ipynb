```python
# pandas
import pandas as pd
```

```python
#1
df = pd.read_csv('auto-mpg.csv')
df.head(10)
```

```python
print(df.shape)
df.describe()
```

```python
# 2
df[(df['model year'] == 75) & (df['weight'] < 3000)][['car name', 'weight', 'mpg']]
```

```python
# 3
df.isna().sum()

```

```python
hp = df[(df['horsepower']!='?')]['horsepower'].astype(int).median()
hps= df['horsepower'].values
hps[hps=='?']=hp
df['horsepower']=hps.astype(int)
df['horsepower']
```

```python
df
```

```python
# 4
df['power_to_weight_ratio'] = df['horsepower'] / df['weight']
df.head()
```

```python
# 5
print(df.groupby('origin')['mpg'].mean())
```

```python
# 6
df_mpg_sort = df.sort_values(by='mpg', ascending=False)
df_mpg_sort.head(10)
```

```python
#7
def performance_score(row): 
    return row['mpg'] * row['acceleration'] / row['weight']
```

```python
df['performance_score'] = df.apply(performance_score,axis =1)
df[['mpg','acceleration','weight','performance_score']].head()
```

```python
# 8
df.groupby('model year')[['mpg', 'weight', 'horsepower']].mean()
```

```python
# 9
df[df['mpg'] > 30][['mpg', 'cylinders', 'horsepower', 'weight']].to_csv('high_mpg_cars.csv', index=False)
```

```python
# 10
Q1, Q3 = df['mpg'].quantile([0.25, 0.75])
IQR = Q3 - Q1
outliers = df[(df['mpg'] < Q1 - 1.5 * IQR) | (df['mpg'] > Q3 + 1.5 * IQR)]
outliers[['car name', 'mpg', 'model year']]
```
